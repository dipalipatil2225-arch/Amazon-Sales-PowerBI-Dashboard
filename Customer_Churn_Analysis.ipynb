```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
```

EDA graph 📊

```python
sns.countplot(x="Churn", data=df)
plt.title("Customer Churn Count")
plt.show()
```

The number of customers who stayed is significantly higher than the number of customers who churned.

```python
sns.countplot(x="gender", hue="Churn", data=df)
plt.title("Gender vs Churn")
plt.show()
```

Customer churn is nearly equal among male and female customers, indicating gender may not strongly influence churn behavior.”

```python
sns.countplot(x="Contract", hue="Churn", data=df)
plt.title("Contract Type vs Churn")
plt.xticks(rotation=20)
plt.show()
```

Customers with month-to-month contracts churn much more, while long-term contract customers are more likely to stay.

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

Customers with higher monthly charges are more likely to churn.

```python
sns.countplot(x="InternetService", hue="Churn", data=df)
plt.title("Internet Service vs Churn")
plt.show()
```

Customers using Fiber Optic internet service have a higher churn rate compared to DSL and customers with no internet service.

```python
sns.countplot(x="SeniorCitizen", hue="Churn", data=df)
plt.title("Senior Citizen vs Churn")
plt.show()
```

Senior citizens show a relatively higher/lower churn trend compared to non-senior customers

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(8,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
